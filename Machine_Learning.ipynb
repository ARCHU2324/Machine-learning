*   **LINEAR_REGRESSION**







1.   Future prediction can be facilitated by utilizing linear regression.
2.   By inputting relevant data variables, linear regression can help to predict outcomes and achieve target variables.
3.  This algorithm operates by analyzing old data, typically in the form of numeric or tabular data.
4.  Linear regression looks for common patterns, learning parameters, and growth trends with fluctuations in order to predict output based on input variables.
5.  As a result, this process is referred to as supervised learning.








```python
#Linear regression

class Linear_reg:
    def start():
        #where x and y is dependent variable(inputs)
        globals()['x']=[]
        globals()['y']=[]
        for i in range(1,6):
            a=int(input("Enter the number of X:"))
            x.append(a)
        for i in range(1,6):
            a=int(input("Enter the number of Y:"))
            y.append(a)
        Linear_reg.start_regression()


    def start_regression():
        m=0
        y_int=0
        counter=0
        globals()['error_list']=[]
        globals()['slope']=[]
        globals()['intercept']=[]
        Linear_reg.reg(m,y_int,counter)

    def reg(m,y_int,counter):
        error=0
        dm=0
        dc=0
        y_hat=[]

        # creating y-hat = mx
        for i in range(len(x)):
            #here we use the slop formula
            #where x is the input , m is slop and the y_int is the intersection
            temp=x[i]*m+y_int
            y_hat.append(temp)
        
        #Calculating error in error and dE/dm in dm and dE/dc in dc
        for i in range(len(x)):
            #y_hat is the actual value and the y is the predicted 
            temp=y_hat[i]-y[i]
            error=error+temp*temp
            dm=temp*x[i]+dm
            dc= dc+temp

        error_list.append(error)
        slope.append(m)
        intercept.append(y_int)
        
        #IF errror is less than 0.05 or it has reached 1000 steps break the fn
        if counter==450:
            
            print("Count reached,Slope is : ",m,"y_int  :",y_int)
            Linear_reg.pred(y_int,m)

        elif error<0.01 :
            
            print("Count reached,Slope is : ",m,"y_int  :",y_int)
            Linear_reg.pred(y_int,m)

        #update the slope and intercept value
        else:
            Linear_reg.update(m,dm,dc,y_int,counter)

    #updating new slope 
    def update(m,dm,dc,y_int,counter):
        '''
        m_new: new weight
        m : old weight
        0.001 value is the learning rate
        dm is the error calculated function
        '''
        m_new=m-0.01*dm
        y_int_new=y_int-0.01*dc
        counter+=1
        Linear_reg.reg(m_new,y_int_new,counter)


    #taking user input value and predicting output
    def pred(y_int,m):
        print("\n Equation of line is :  \t Y = {} * X   +  ( {} )\n".format(m,y_int))
        x=int(input("Enter the number you want to predict:"))
        ans= m*x+y_int
        return print("\n Answer is :",ans)
```
